### Machine Learning
For this next section, we'll be implementing Random Forests to try and find gene interaction networks that can predict outcome. For this to work, we must aggregate based on genes. 

```python
def mutation_present(dataframe_genes, patient_id, gene):
    query_result = query_by_gene(dataframe_genes, gene)
    result = len(query_result[query_result['PATIENT_ID']== patient_id]) > 0
    return 0 if result == 0 else 1
```

```python
patients_gene_network = pd.DataFrame()
patients_gene_network["PATIENT_ID"] = df_clinical_patients["PATIENT_ID"]
```

```python
clinical_data_with_patients["Hugo_Symbol"]
```

```python
all_genes_mapped = list(set(clinical_data_with_patients["Hugo_Symbol"]))
```

```python
def create_gene_map(dataframe_genes, all_genes):
    gene_map = {}
    for gene in all_genes:
        patient_list = list(dataframe_genes[dataframe_genes['Hugo_Symbol']==gene]["PATIENT_ID"])
        gene_map[gene] = patient_list
    return gene_map
```

```python
gene_map = create_gene_map(clinical_data_with_patients, all_genes_mapped)
```

```python
def add_patient_gene_network(dataframe, mapped_genes, gene):
    temp_dataframe = dataframe.copy()
    patients = mapped_genes[gene]
    truth_value = lambda row: row["PATIENT_ID"] in patients
    label = lambda row:  True if truth_value(row) else False
    temp_dataframe[gene] = temp_dataframe.apply(label, axis="columns") #Apply Label
    return temp_dataframe
```

```python

for gene in all_genes_mapped:
    patients_gene_network = add_patient_gene_network(patients_gene_network, gene_map, gene)
patients_gene_network
```

```python
def add_class(dataframe, alive_patients_list):
    temp_dataframe = dataframe.copy()
    truth_value = lambda row: row["PATIENT_ID"] in alive_patients_list
    label = lambda row:  1 if truth_value(row) else 0
    temp_dataframe["Survived"] = temp_dataframe.apply(label, axis="columns") #Apply Label
    return temp_dataframe
```

```python
def create_table_of_analysis(dataframe, real_label_column_name, test_label_column_name, positive_result, negative_result):
    row_count = dataframe.shape[0]
    true_positive = dataframe.query('`{0}` == `{1}` and `{1}` == "{2}"'.format(real_label_column_name, test_label_column_name, positive_result))
    false_positive = dataframe.query('`{0}` == "{3}" and `{1}` == "{2}"'.format(real_label_column_name, test_label_column_name, positive_result, negative_result))
    true_negative = dataframe.query('`{0}` == `{1}` and `{0}` == "{2}"'.format(real_label_column_name, test_label_column_name, negative_result))
    false_negative = dataframe.query('`{0}` == "{2}" and `{1}` == "{3}"'.format(real_label_column_name, test_label_column_name, positive_result, negative_result))

    true_positive_r = true_positive.shape[0]
    false_positive_r = false_positive.shape[0]
    true_negative_r = true_negative.shape[0]
    false_negative_r = false_negative.shape[0]
    accuracy = float((true_positive_r + true_negative_r)) / float(row_count)

    tpr = (float(true_positive_r) / float(true_positive_r + false_negative_r)
           if (true_positive_r + false_negative_r) > 0 else 0.0)
    tnr = (float(true_negative_r) / float(true_negative_r + false_positive_r)
           if (true_negative_r + false_positive_r) > 0 else 0.0)

    d = {"TP": [true_positive_r], "FP": [false_positive_r], "TN": [true_negative_r],
         "FN": [false_negative_r], "Accuracy": [accuracy], "TPR": [tpr], "TNR": [tnr]}

    return pd.DataFrame(data=d)
```

```python
max_accuracy = -1
df4 = pd.DataFrame(data = X_test, columns = x_headers)
d = {"Number of Trees": [np.nan], "Depth":[np.nan], "TP":[np.nan],"FP":[np.nan],"TN":[np.nan],"FN":[np.nan], "Accuracy": [np.nan], "TPR":[np.nan], "TNR": [np.nan]}

table_of_results = pd.DataFrame(data=d)

    
for total_trees in range(1,11):
    for depth in range(1,6):
        temp_dataframe = df4.copy()
        clf_model = RandomForestClassifier(max_depth=depth, n_estimators=total_trees, random_state=0) 
        clf_model.fit(X_train,y_train)

        y_predicted = clf_model.predict(X_test)
        

        temp_dataframe["Class"] = Y_test
        temp_dataframe["Predicted"] = y_predicted
        #print("Number of Trees - ", total_trees)
        #print("Overall Depth - ", depth)
        table_accuracy = create_table_of_analysis(temp_dataframe, "Class", "Predicted", 1, 0)
        table_accuracy["Number of Trees"] = total_trees
        table_accuracy["Depth"] = depth
        accuracy = table_accuracy.iloc[0]["Accuracy"]
        table_of_results = pd.concat([table_of_results,table_accuracy])
        
        if accuracy > max_accuracy:
            maximum_table = table_accuracy
            optimal_features = {"Total Trees" : total_trees, "Depth": depth}
            maximum_model = clf_model
            maximum_confusion_matrix = confusion_matrix(Y_test, y_predicted)
            max_accuracy = accuracy
```

### Association Algorithm
For Nominal Association, I used Association from Scipy.

The default is Cramer's V

>Cramer’s V, Tschuprow’s T and Pearson’s Contingency Coefficient, all measure the degree to which two nominal or ordinal variables are related, or the level of their association. This differs from correlation, although many often mistakenly consider them equivalent. Correlation measures in what way two variables are related, whereas, association measures how related the variables are. As such, association does not subsume independent variables, and is rather a test of independence. A value of 1.0 indicates perfect association, and 0.0 means the variables have no association.

Cramer's V:
>φc is the intercorrelation of two discrete variables[2] and may be used with variables having two or more levels. φc is a symmetrical measure: it does not matter which variable we place in the columns and which in the rows. Also, the order of rows/columns doesn't matter, so φc may be used with nominal data types or higher (notably, ordered or numerical).
https://en.wikipedia.org/wiki/Cram%C3%A9r%27s_V

I also used Tschuprow's T
>In statistics, Tschuprow's T is a measure of association between two nominal variables, giving a value between 0 and 1 (inclusive). It is closely related to Cramér's V, coinciding with it for square contingency tables. It was published by Alexander Tschuprow (alternative spelling: Chuprov) in 1939.[1]

Contingency Tables are defined as follows
>In statistics, a contingency table (also known as a cross tabulation or crosstab) is a type of table in a matrix format that displays the (multivariate) frequency distribution of the variables. They are heavily used in survey research, business intelligence, engineering, and scientific research. They provide a basic picture of the interrelation between two variables and can help find interactions between them. ## Part 1 — Multi-Run Gene Selection Frequency & Part 2 — Pairwise Cramer's V

We run the Random Forest pipeline **100 times** with varied random seeds to
measure which genes are *consistently* chosen as decision nodes, not just in
one lucky split.

**Part 1:** Each run fits a fresh RF (grid-searched over n_estimators 2-10,
max_depth 2-5) on a 50/50 balanced split. Genes found in any decision node
are recorded. After 100 runs we rank genes by how often they appeared.

**Part 2:** Within each run, we compute **Cramer's V** for every pair of
selected genes using the test-set contingency table (gene presence 2/1 encoded).
After 100 runs we aggregate co-occurrence counts and mean Cramer's V per pair.

```python
import re
import numpy as np
import pandas as pd
from collections import defaultdict
from itertools import combinations
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree as sk_tree
from scipy.stats.contingency import association as cramers_association

N_OBS_RUNS = 100
GENE_RE = re.compile(r'([A-Z])\w+[0-9]*')

# Use the class-balanced dataset and the same feature list as the existing model
balanced_df = new_df.copy().reset_index(drop=True)
feat_cols = list(curious_genes)  # gene column names used in existing RF

# Storage
obs_gene_counts  = defaultdict(int)    # gene -> runs selected in
obs_pair_counts  = defaultdict(int)    # (ga,gb) -> co-occurrence runs
obs_pair_cramers = defaultdict(list)   # (ga,gb) -> [cramers_v per run]
all_obs_sets     = []                  # list of frozensets, one per run

def _extract_genes(model, feature_names):
    found = set()
    for est in model.estimators_:
        text = sk_tree.export_text(est, feature_names=list(feature_names))
        for line in text.split('\n'):
            mo = GENE_RE.search(line)
            if mo:
                found.add(mo.group())
    return found

def _encode_21(df_subset):
    return df_subset.astype(int).add(1)   # False/0 -> 1, True/1 -> 2

print("Starting 100 observed runs...")
for i in range(N_OBS_RUNS):
    rs = (i + 1) * 7

    X = balanced_df[feat_cols]
    y = balanced_df['Survived']
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.5, random_state=rs)

    # Grid search
    best_acc, best_model = 0, None
    for n_est in range(2, 11):
        for depth in range(2, 6):
            clf = RandomForestClassifier(n_estimators=n_est, max_depth=depth,
                                         random_state=rs)
            clf.fit(X_train, y_train)
            acc = clf.score(X_test, y_test)
            if acc > best_acc:
                best_acc, best_model = acc, clf

    run_genes = _extract_genes(best_model, feat_cols)
    all_obs_sets.append(frozenset(run_genes))
    for g in run_genes:
        obs_gene_counts[g] += 1

    # Cramer's V for every pair in this run (test set, 2/1 encoded)
    genes_sorted = sorted(run_genes)
    if len(genes_sorted) >= 2:
        test_enc = _encode_21(X_test[genes_sorted])
        for ga, gb in combinations(genes_sorted, 2):
            pair = tuple(sorted([ga, gb]))
            obs_pair_counts[pair] += 1
            ct = pd.crosstab(test_enc[ga], test_enc[gb])
            if ct.shape == (2, 2):
                try:
                    v = cramers_association(ct.values, method='cramer')
                    if v > 0:
                        obs_pair_cramers[pair].append(v)
                except Exception:
                    pass

    if (i + 1) % 25 == 0:
        print(f"  Observed run {i+1}/{N_OBS_RUNS} done  (best_acc={best_acc:.3f})")

print("\nParts 1 & 2 complete.")
```

```python
# Part 1 results — gene selection frequency table
gene_freq_df = pd.DataFrame([
    {'Gene': g, 'Observed_Count': c, 'Observed_Rate': c / N_OBS_RUNS}
    for g, c in obs_gene_counts.items()
]).sort_values('Observed_Count', ascending=False).reset_index(drop=True)

print(f"=== Part 1: Gene Selection Frequency over {N_OBS_RUNS} observed runs ===")
print(gene_freq_df.to_string(index=False, float_format='%.2f'))
```

```python
# Part 2 results — pairwise co-occurrence and mean Cramer's V
pair_rows = []
for pair, cnt in obs_pair_counts.items():
    v_list = obs_pair_cramers.get(pair, [])
    pair_rows.append({
        'Gene_A': pair[0],
        'Gene_B': pair[1],
        'Obs_Cooccurrence': cnt,
        'Obs_Rate': cnt / N_OBS_RUNS,
        'Mean_CramersV': np.mean(v_list) if v_list else 0.0,
        'N_V_Obs': len(v_list)
    })

obs_pairs_df = pd.DataFrame(pair_rows).sort_values('Obs_Cooccurrence', ascending=False).reset_index(drop=True)

print(f"=== Part 2: Pairwise Co-occurrence & Mean Cramer's V ({N_OBS_RUNS} observed runs) ===")
print(obs_pairs_df.head(20).to_string(index=False, float_format='%.3f'))
```

## Part 3 — Permutation Null Model (100 Runs)

We run the identical pipeline 100 more times, but before each run the
`Survived` column is **randomly shuffled**, destroying any real relationship
between gene fusions and survival outcome. This builds an empirical null
distribution: how often would a gene (or pair) appear purely by chance,
given its base-rate mutation frequency in the dataset?

```python
N_PERM_RUNS = 100

perm_gene_counts  = defaultdict(int)
perm_pair_counts  = defaultdict(int)
perm_pair_cramers = defaultdict(list)
all_perm_sets     = []

print("Starting 100 permuted runs...")
for i in range(N_PERM_RUNS):
    rs = (i + 1) * 7

    # Shuffle Survived — destroys signal
    perm_survived = np.random.RandomState(seed=i + 100).permutation(
        balanced_df['Survived'].values)
    perm_df = balanced_df.copy()
    perm_df['Survived'] = perm_survived

    X = perm_df[feat_cols]
    y = perm_df['Survived']
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.5, random_state=rs)

    best_acc, best_model = 0, None
    for n_est in range(2, 11):
        for depth in range(2, 6):
            clf = RandomForestClassifier(n_estimators=n_est, max_depth=depth,
                                         random_state=rs)
            clf.fit(X_train, y_train)
            acc = clf.score(X_test, y_test)
            if acc > best_acc:
                best_acc, best_model = acc, clf

    run_genes = _extract_genes(best_model, feat_cols)
    all_perm_sets.append(frozenset(run_genes))
    for g in run_genes:
        perm_gene_counts[g] += 1

    genes_sorted = sorted(run_genes)
    if len(genes_sorted) >= 2:
        test_enc = _encode_21(X_test[genes_sorted])
        for ga, gb in combinations(genes_sorted, 2):
            pair = tuple(sorted([ga, gb]))
            perm_pair_counts[pair] += 1
            ct = pd.crosstab(test_enc[ga], test_enc[gb])
            if ct.shape == (2, 2):
                try:
                    v = cramers_association(ct.values, method='cramer')
                    if v > 0:
                        perm_pair_cramers[pair].append(v)
                except Exception:
                    pass

    if (i + 1) % 25 == 0:
        print(f"  Permuted run {i+1}/{N_PERM_RUNS} done  (best_acc={best_acc:.3f})")

print("\nPart 3 complete.")
```

## Part 4 — Two-Tier Statistical Comparison (Observed vs Permuted)

**Tier 1 (gene level):** For each gene, compare observed selection frequency
against its permuted null. The empirical p-value is the fraction of permuted
runs where the gene was selected — if the null is true, permuted and observed
rates should be equal.

**Tier 2 (pair level):** Same logic applied to gene pairs that appeared in at
least 3 observed runs. A pair with a low p-value and ratio > 1.5 is co-occurring
more than its base-rate frequencies predict.

```python
# Raw fusion frequency from the full (unbalanced) patient-gene network
all_genes_seen = set(obs_gene_counts) | set(perm_gene_counts)
n_total_patients = len(patients_gene_network)
raw_freq_map = {}
for g in all_genes_seen:
    if g in patients_gene_network.columns:
        raw_freq_map[g] = float(patients_gene_network[g].astype(int).sum()) / n_total_patients
    else:
        raw_freq_map[g] = np.nan

# ── Tier 1: Gene level ─────────────────────────────────────────────────────────
tier1_rows = []
for g in all_genes_seen:
    obs_c   = obs_gene_counts.get(g, 0)
    perm_c  = perm_gene_counts.get(g, 0)
    obs_r   = obs_c  / N_OBS_RUNS
    perm_r  = perm_c / N_PERM_RUNS
    ratio   = obs_r / perm_r if perm_r > 0 else (np.inf if obs_r > 0 else 1.0)
    # p-value: fraction of permuted runs that selected gene >= obs_r
    p_val   = perm_c / N_PERM_RUNS   # = perm_rate (each perm run is one Bernoulli trial)
    tier1_rows.append({
        'Gene': g,
        'Obs_Freq': obs_c,
        'Perm_Freq': perm_c,
        'Obs_Rate': obs_r,
        'Perm_Rate': perm_r,
        'Ratio': ratio,
        'P_Value': p_val,
        'Raw_Fusion_Freq': raw_freq_map.get(g, np.nan),
        'Tier1_Signal': (p_val < 0.05) and (ratio > 1.5)
    })

tier1_df = pd.DataFrame(tier1_rows).sort_values('Obs_Freq', ascending=False).reset_index(drop=True)
print("=== Tier 1: Gene-Level Comparison ===")
print(tier1_df.to_string(index=False, float_format='%.3f'))
print(f"\nTier 1 signal genes (p<0.05, ratio>1.5): {tier1_df.Tier1_Signal.sum()}")

# ── Tier 2: Pair level ─────────────────────────────────────────────────────────
all_pairs = set(obs_pair_counts) | set(perm_pair_counts)
tier2_rows = []
for pair in all_pairs:
    obs_c  = obs_pair_counts.get(pair, 0)
    if obs_c < 3:
        continue
    perm_c = perm_pair_counts.get(pair, 0)
    obs_r  = obs_c  / N_OBS_RUNS
    perm_r = perm_c / N_PERM_RUNS
    ratio  = obs_r / perm_r if perm_r > 0 else (np.inf if obs_r > 0 else 1.0)
    p_val  = perm_c / N_PERM_RUNS
    v_list = obs_pair_cramers.get(pair, [])
    tier2_rows.append({
        'Gene_A': pair[0],
        'Gene_B': pair[1],
        'Obs_Freq': obs_c,
        'Perm_Freq': perm_c,
        'Obs_Rate': obs_r,
        'Perm_Rate': perm_r,
        'Ratio': ratio,
        'P_Value': p_val,
        'Mean_CramersV': np.mean(v_list) if v_list else 0.0,
        'Tier2_Signal': (p_val < 0.05) and (ratio > 1.5)
    })

tier2_df = pd.DataFrame(tier2_rows).sort_values('Obs_Freq', ascending=False).reset_index(drop=True)
print("\n=== Tier 2: Pair-Level Comparison (Obs_Freq >= 3) ===")
print(tier2_df.to_string(index=False, float_format='%.3f'))
print(f"\nTier 2 signal pairs (p<0.05, ratio>1.5): {tier2_df.Tier2_Signal.sum()}")
```

## Part 5 — Near-Significant Candidates (p < 0.10)

We relax the threshold to **p < 0.10** to surface gene pairs whose
co-occurrence is beginning to emerge above the permutation null. These are
candidates for follow-up, not confirmed findings — but in a dataset of
this size, they represent the clearest biological leads the data can offer.
