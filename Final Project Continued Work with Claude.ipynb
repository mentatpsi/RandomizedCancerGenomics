### Machine Learning
For this next section, we'll be implementing Random Forests to try and find gene interaction networks that can predict outcome. For this to work, we must aggregate based on genes. 

```python
def mutation_present(dataframe_genes, patient_id, gene):
    query_result = query_by_gene(dataframe_genes, gene)
    result = len(query_result[query_result['PATIENT_ID']== patient_id]) > 0
    return 0 if result == 0 else 1
```

```python
patients_gene_network = pd.DataFrame()
patients_gene_network["PATIENT_ID"] = df_clinical_patients["PATIENT_ID"]
```

```python
all_genes_mapped = list(set(clinical_data_with_patients["Hugo_Symbol"]))
```

```python
def create_gene_map(dataframe_genes, all_genes):
    gene_map = {}
    for gene in all_genes:
        patient_list = list(dataframe_genes[dataframe_genes['Hugo_Symbol']==gene]["PATIENT_ID"])
        gene_map[gene] = patient_list
    return gene_map
```

```python
gene_map = create_gene_map(clinical_data_with_patients, all_genes_mapped)
```

```python
def add_patient_gene_network(dataframe, mapped_genes, gene):
    temp_dataframe = dataframe.copy()
    patients = mapped_genes[gene]
    truth_value = lambda row: row["PATIENT_ID"] in patients
    label = lambda row:  True if truth_value(row) else False
    temp_dataframe[gene] = temp_dataframe.apply(label, axis="columns") #Apply Label
    return temp_dataframe
```

```python

for gene in all_genes_mapped:
    patients_gene_network = add_patient_gene_network(patients_gene_network, gene_map, gene)
patients_gene_network
```

```python
def create_table_of_analysis(dataframe, real_label_column_name, test_label_column_name, positive_result, negative_result):
    row_count = dataframe.shape[0]
    true_positive = dataframe.query('`{0}` == `{1}` and `{1}` == "{2}"'.format(real_label_column_name, test_label_column_name, positive_result))
    false_positive = dataframe.query('`{0}` == "{3}" and `{1}` == "{2}"'.format(real_label_column_name, test_label_column_name, positive_result, negative_result))
    true_negative = dataframe.query('`{0}` == `{1}` and `{0}` == "{2}"'.format(real_label_column_name, test_label_column_name, negative_result))
    false_negative = dataframe.query('`{0}` == "{2}" and `{1}` == "{3}"'.format(real_label_column_name, test_label_column_name, positive_result, negative_result))

    true_positive_r = true_positive.shape[0]
    false_positive_r = false_positive.shape[0]
    true_negative_r = true_negative.shape[0]
    false_negative_r = false_negative.shape[0]
    accuracy = float((true_positive_r + true_negative_r)) / float(row_count)

    tpr = (float(true_positive_r) / float(true_positive_r + false_negative_r)
           if (true_positive_r + false_negative_r) > 0 else 0.0)
    tnr = (float(true_negative_r) / float(true_negative_r + false_positive_r)
           if (true_negative_r + false_positive_r) > 0 else 0.0)

    d = {"TP": [true_positive_r], "FP": [false_positive_r], "TN": [true_negative_r],
         "FN": [false_negative_r], "Accuracy": [accuracy], "TPR": [tpr], "TNR": [tnr]}

    return pd.DataFrame(data=d)
```

### Association Algorithm
For Nominal Association, I used Association from Scipy.

The default is Cramer's V

>Cramer’s V, Tschuprow’s T and Pearson’s Contingency Coefficient, all measure the degree to which two nominal or ordinal variables are related, or the level of their association. This differs from correlation, although many often mistakenly consider them equivalent. Correlation measures in what way two variables are related, whereas, association measures how related the variables are. As such, association does not subsume independent variables, and is rather a test of independence. A value of 1.0 indicates perfect association, and 0.0 means the variables have no association.

Cramer's V:
>φc is the intercorrelation of two discrete variables[2] and may be used with variables having two or more levels. φc is a symmetrical measure: it does not matter which variable we place in the columns and which in the rows. Also, the order of rows/columns doesn't matter, so φc may be used with nominal data types or higher (notably, ordered or numerical).
https://en.wikipedia.org/wiki/Cram%C3%A9r%27s_V

I also used Tschuprow's T
>In statistics, Tschuprow's T is a measure of association between two nominal variables, giving a value between 0 and 1 (inclusive). It is closely related to Cramér's V, coinciding with it for square contingency tables. It was published by Alexander Tschuprow (alternative spelling: Chuprov) in 1939.[1]

Contingency Tables are defined as follows
>In statistics, a contingency table (also known as a cross tabulation or crosstab) is a type of table in a matrix format that displays the (multivariate) frequency distribution of the variables. They provide a basic picture of the interrelation between two variables and can help find interactions between them. ## Extended Analysis — May 2026

The original pipeline identified candidate risk genes through set-theoretic filtering and trained a Random Forest classifier on a binary gene-presence matrix. That approach has two limitations: (1) a single classifier run is non-deterministic — the random seed, bootstrap sampling, and feature subsampling at each split all introduce variance, so a gene important in one run may not appear in the next; and (2) it provides no statistical baseline for distinguishing genuine survival signal from genes that appear frequently simply because they are common in this cohort.

This extended analysis addresses both limitations through three additions:

1. **Stability analysis (Parts 1–2):** The Random Forest pipeline is run 100 times with varied random seeds. Genes are ranked by how consistently the model selects them as decision nodes — separating genuine signal from fortunate splits.

2. **Permutation null model (Part 3):** The same 100-run pipeline is repeated with the survival label randomly shuffled, producing an empirical null distribution. This answers: *how often would this gene appear purely by chance, given its base rate in this dataset?*

3. **Two-tier statistical framework (Parts 4–6):** Observed co-occurrence rates are compared against the permutation null to compute empirical p-values, followed by Cramer's V chi-square testing (Part 7) and Fisher's exact test validation of key findings (Part 8).

**Data source:** MSK Breast Cancer 2018 cohort (`breast_msk_2018`), cBioPortal. n = 1,918 patients sequenced with the MSK-IMPACT 410-gene panel (Zehir et al., *Nature Medicine*, 2017).

## Part 1 — Gene Selection Stability Across 100 Random Forest Runs

### Why a Single Run Is Not Enough

A Random Forest is an ensemble of decision trees, each grown on a bootstrap sample of the training data with a random subset of features available at every split (Breiman, 2001). This randomization reduces variance relative to a single tree, but the *genes* selected as split criteria still vary across runs. A gene that appears important in one forest may not appear in the next if it was excluded from the random feature subset at the critical split node.

To measure which genes are *reliably* important — not just important in one run — we repeat the pipeline 100 times with varied random seeds and record how often each gene appears as a split criterion across all runs. This is analogous to the **stability selection** framework of Meinshausen & Bühlmann (2010), which uses subsampling to assess which features a model consistently relies upon.

### Implementation

Each of the 100 runs:
1. Splits the class-balanced dataset 50/50 into train and test sets (preserving the balanced class distribution).
2. Grid-searches over `n_estimators` ∈ {2…10} and `max_depth` ∈ {2…5}, selecting the hyperparameter combination with the highest test-set accuracy.
3. Exports every decision tree's text representation and extracts gene names from split lines using a regular expression — genes that the forest actually used to partition patients.

After 100 runs, each gene receives a **selection rate**: the fraction of runs in which it appeared in at least one decision node. A gene with rate 0.80 was chosen by 80 of 100 forests.

## Part 2 — Pairwise Cramer's V Across Runs

### Cramer's V as an Association Measure

Within each run, we compute **Cramer's V** for every pair of genes the forest selected, using the test-set contingency table. Cramer's V is a symmetric measure of association between two nominal (categorical) variables derived from the Pearson chi-square statistic (Cramér, 1946):

$$V = \sqrt{\frac{\chi^2}{n \cdot (k-1)}}$$

where *n* is the sample size and *k* = min(rows, columns) of the contingency table (= 2 for binary gene presence). V ranges from 0 (independence) to 1 (perfect association) and, unlike the raw χ², is not inflated by sample size.

Aggregating Cramer's V across 100 runs gives a **mean association score** per gene pair that is more stable than any single-run estimate and less sensitive to the particular random seed.

### References
- Breiman, L. (2001). Random Forests. *Machine Learning*, 45(1), 5–32.
- Cramér, H. (1946). *Mathematical Methods of Statistics*. Princeton University Press.
- Meinshausen, N. & Bühlmann, P. (2010). Stability selection. *Journal of the Royal Statistical Society: Series B*, 72(4), 417–473.

```python
import re
import numpy as np
import pandas as pd
from collections import defaultdict
from itertools import combinations
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree as sk_tree
from scipy.stats.contingency import association as cramers_association

N_OBS_RUNS = 100
GENE_RE = re.compile(r'([A-Z])\w+[0-9]*')

# Use the class-balanced dataset and the same feature list as the existing model
balanced_df = new_df.copy().reset_index(drop=True)
feat_cols = list(curious_genes)  # gene column names used in existing RF

# Storage
obs_gene_counts  = defaultdict(int)    # gene -> runs selected in
obs_pair_counts  = defaultdict(int)    # (ga,gb) -> co-occurrence runs
obs_pair_cramers = defaultdict(list)   # (ga,gb) -> [cramers_v per run]
all_obs_sets     = []                  # list of frozensets, one per run

def _extract_genes(model, feature_names):
    found = set()
    for est in model.estimators_:
        text = sk_tree.export_text(est, feature_names=list(feature_names))
        for line in text.split('\n'):
            mo = GENE_RE.search(line)
            if mo:
                found.add(mo.group())
    return found

def _encode_21(df_subset):
    return df_subset.astype(int).add(1)   # False/0 -> 1, True/1 -> 2

print("Starting 100 observed runs...")
for i in range(N_OBS_RUNS):
    rs = (i + 1) * 7

    X = balanced_df[feat_cols]
    y = balanced_df['Survived']
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.5, random_state=rs)

    # Grid search
    best_acc, best_model = 0, None
    for n_est in range(2, 11):
        for depth in range(2, 6):
            clf = RandomForestClassifier(n_estimators=n_est, max_depth=depth,
                                         random_state=rs)
            clf.fit(X_train, y_train)
            acc = clf.score(X_test, y_test)
            if acc > best_acc:
                best_acc, best_model = acc, clf

    run_genes = _extract_genes(best_model, feat_cols)
    all_obs_sets.append(frozenset(run_genes))
    for g in run_genes:
        obs_gene_counts[g] += 1

    # Cramer's V for every pair in this run (test set, 2/1 encoded)
    genes_sorted = sorted(run_genes)
    if len(genes_sorted) >= 2:
        test_enc = _encode_21(X_test[genes_sorted])
        for ga, gb in combinations(genes_sorted, 2):
            pair = tuple(sorted([ga, gb]))
            obs_pair_counts[pair] += 1
            ct = pd.crosstab(test_enc[ga], test_enc[gb])
            if ct.shape == (2, 2):
                try:
                    v = cramers_association(ct.values, method='cramer')
                    if v > 0:
                        obs_pair_cramers[pair].append(v)
                except Exception:
                    pass

    if (i + 1) % 25 == 0:
        print(f"  Observed run {i+1}/{N_OBS_RUNS} done  (best_acc={best_acc:.3f})")

print("\nParts 1 & 2 complete.")
```

## Part 3 — Permutation Null Model (100 Runs)

### The Problem of Base-Rate Confounding

Gene selection rates from Part 1 cannot distinguish between two explanations:
1. The gene is genuinely associated with survival and the model is learning a real signal.
2. The gene is frequently fused in this cohort and the model selects it because it has the highest variance to split on — regardless of whether it relates to survival.

A gene fused in 40% of patients will appear in almost every tree simply because it is the most informative feature *on variance grounds alone*. We need a baseline that accounts for each gene's mutation frequency before we can claim survival-specific enrichment.

### Permutation Testing

**Permutation testing** addresses this by constructing an empirical null distribution — a distribution of outcomes we would expect if no real signal existed (Good, 2005). The procedure:

1. **Shuffle the `Survived` label** randomly across all patients. This destroys any true relationship between gene fusions and outcome while preserving everything else:
   - Each gene's marginal mutation frequency (base rate) is unchanged.
   - The correlation structure among genes (which genes co-occur in the same patients) is unchanged.
   - The class balance of the dataset is unchanged.

2. Run the identical 100-run Random Forest pipeline on the shuffled data.

Under the null hypothesis that gene fusions carry no survival information, permuted and observed selection rates should be statistically indistinguishable. Any gene selected more often in the observed runs than in the permuted runs is enriched *above its base-rate null* — the excess is attributable to survival association, not mutation frequency alone.

### Statistical Note on Permutation P-Values

With 100 permuted runs, the minimum resolvable p-value is 1/100 = 0.01. A reported p = 0.00 should be interpreted as p < 0.01, not as exactly zero — the resolution is limited by the number of permutations. More permutations (e.g. 1,000) would give finer resolution at proportionally higher computational cost (Phipson & Smyth, 2010).

### References
- Good, P.I. (2005). *Permutation, Parametric, and Bootstrap Tests of Hypotheses* (3rd ed.). Springer.
- Phipson, B. & Smyth, G.K. (2010). Permutation p-values should never be zero: calculating exact p-values when permutations are randomly drawn. *Statistical Applications in Genetics and Molecular Biology*, 9(1), Article 39.

```python
N_PERM_RUNS = 100

perm_gene_counts  = defaultdict(int)
perm_pair_counts  = defaultdict(int)
perm_pair_cramers = defaultdict(list)
all_perm_sets     = []

print("Starting 100 permuted runs...")
for i in range(N_PERM_RUNS):
    rs = (i + 1) * 7

    # Shuffle Survived — destroys signal
    perm_survived = np.random.RandomState(seed=i + 100).permutation(
        balanced_df['Survived'].values)
    perm_df = balanced_df.copy()
    perm_df['Survived'] = perm_survived

    X = perm_df[feat_cols]
    y = perm_df['Survived']
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.5, random_state=rs)

    best_acc, best_model = 0, None
    for n_est in range(2, 11):
        for depth in range(2, 6):
            clf = RandomForestClassifier(n_estimators=n_est, max_depth=depth,
                                         random_state=rs)
            clf.fit(X_train, y_train)
            acc = clf.score(X_test, y_test)
            if acc > best_acc:
                best_acc, best_model = acc, clf

    run_genes = _extract_genes(best_model, feat_cols)
    all_perm_sets.append(frozenset(run_genes))
    for g in run_genes:
        perm_gene_counts[g] += 1

    genes_sorted = sorted(run_genes)
    if len(genes_sorted) >= 2:
        test_enc = _encode_21(X_test[genes_sorted])
        for ga, gb in combinations(genes_sorted, 2):
            pair = tuple(sorted([ga, gb]))
            perm_pair_counts[pair] += 1
            ct = pd.crosstab(test_enc[ga], test_enc[gb])
            if ct.shape == (2, 2):
                try:
                    v = cramers_association(ct.values, method='cramer')
                    if v > 0:
                        perm_pair_cramers[pair].append(v)
                except Exception:
                    pass

    if (i + 1) % 25 == 0:
        print(f"  Permuted run {i+1}/{N_PERM_RUNS} done  (best_acc={best_acc:.3f})")

print("\nPart 3 complete.")
```

## Part 4 — Two-Tier Statistical Comparison (Observed vs Permuted)

We now compare observed gene and gene-pair frequencies against their permutation null distributions to compute empirical p-values and enrichment ratios.

### Tier 1 — Gene Level

For each gene, two quantities are computed:

- **Empirical p-value** = the fraction of permuted runs in which the gene was selected. This estimates the probability of observing a selection rate at least as high as the null, under the assumption that no real survival signal exists. A gene selected in 3 of 100 permuted runs has an empirical p-value of 0.03.

- **Enrichment ratio** = observed rate / permuted rate. A ratio of 2.0 means the gene was selected twice as often in the real data as in the shuffled null. Ratios below 1.0 mean the gene is *less* selected in real data than expected from its mutation frequency alone — it may be an artefact of class balance rather than survival signal.

A gene satisfying both **p < 0.05** and **ratio > 1.5** is flagged as a Tier 1 signal gene: its selection by the Random Forest exceeds what base-rate mutation frequency alone predicts.

### Tier 2 — Pair Level

The same logic is applied to gene pairs that co-appeared in at least **3 observed runs** (a minimum frequency filter to exclude singleton noise). A pair with a low empirical p-value and ratio > 1.5 is co-occurring in the Random Forest trees more than the permuted null predicts — suggesting that the two genes are not merely common in the cohort, but that the model is learning to split on them together because their joint presence is informative for survival classification.

### Design Rationale

The two-tier structure separates two distinct questions:
- **Tier 1** asks: *is this gene enriched in the survival-predictive model above its base-rate frequency?*
- **Tier 2** asks: *is this gene pair co-enriched beyond what the individual gene frequencies would predict?*

A pair can fail Tier 2 (both genes are selected at base rate, but they happen to co-occur) while the individual genes pass Tier 1. Conversely, a pair can pass Tier 2 (co-occurrence is enriched) even when individual gene selection rates are modest.

```python
# Raw fusion frequency from the full (unbalanced) patient-gene network
all_genes_seen = set(obs_gene_counts) | set(perm_gene_counts)
n_total_patients = len(patients_gene_network)
raw_freq_map = {}
for g in all_genes_seen:
    if g in patients_gene_network.columns:
        raw_freq_map[g] = float(patients_gene_network[g].astype(int).sum()) / n_total_patients
    else:
        raw_freq_map[g] = np.nan

# ── Tier 1: Gene level ─────────────────────────────────────────────────────────
tier1_rows = []
for g in all_genes_seen:
    obs_c   = obs_gene_counts.get(g, 0)
    perm_c  = perm_gene_counts.get(g, 0)
    obs_r   = obs_c  / N_OBS_RUNS
    perm_r  = perm_c / N_PERM_RUNS
    ratio   = obs_r / perm_r if perm_r > 0 else (np.inf if obs_r > 0 else 1.0)
    # p-value: fraction of permuted runs that selected gene >= obs_r
    p_val   = perm_c / N_PERM_RUNS   # = perm_rate (each perm run is one Bernoulli trial)
    tier1_rows.append({
        'Gene': g,
        'Obs_Freq': obs_c,
        'Perm_Freq': perm_c,
        'Obs_Rate': obs_r,
        'Perm_Rate': perm_r,
        'Ratio': ratio,
        'P_Value': p_val,
        'Raw_Fusion_Freq': raw_freq_map.get(g, np.nan),
        'Tier1_Signal': (p_val < 0.05) and (ratio > 1.5)
    })

tier1_df = pd.DataFrame(tier1_rows).sort_values('Obs_Freq', ascending=False).reset_index(drop=True)
print("=== Tier 1: Gene-Level Comparison ===")
print(tier1_df.to_string(index=False, float_format='%.3f'))
print(f"\nTier 1 signal genes (p<0.05, ratio>1.5): {tier1_df.Tier1_Signal.sum()}")

# ── Tier 2: Pair level ─────────────────────────────────────────────────────────
all_pairs = set(obs_pair_counts) | set(perm_pair_counts)
tier2_rows = []
for pair in all_pairs:
    obs_c  = obs_pair_counts.get(pair, 0)
    if obs_c < 3:
        continue
    perm_c = perm_pair_counts.get(pair, 0)
    obs_r  = obs_c  / N_OBS_RUNS
    perm_r = perm_c / N_PERM_RUNS
    ratio  = obs_r / perm_r if perm_r > 0 else (np.inf if obs_r > 0 else 1.0)
    p_val  = perm_c / N_PERM_RUNS
    v_list = obs_pair_cramers.get(pair, [])
    tier2_rows.append({
        'Gene_A': pair[0],
        'Gene_B': pair[1],
        'Obs_Freq': obs_c,
        'Perm_Freq': perm_c,
        'Obs_Rate': obs_r,
        'Perm_Rate': perm_r,
        'Ratio': ratio,
        'P_Value': p_val,
        'Mean_CramersV': np.mean(v_list) if v_list else 0.0,
        'Tier2_Signal': (p_val < 0.05) and (ratio > 1.5)
    })

tier2_df = pd.DataFrame(tier2_rows).sort_values('Obs_Freq', ascending=False).reset_index(drop=True)
print("\n=== Tier 2: Pair-Level Comparison (Obs_Freq >= 3) ===")
print(tier2_df.to_string(index=False, float_format='%.3f'))
print(f"\nTier 2 signal pairs (p<0.05, ratio>1.5): {tier2_df.Tier2_Signal.sum()}")
```

## Part 5 — Near-Significant Candidates (p < 0.10)

### Threshold Choice in Exploratory Genomics

In confirmatory hypothesis testing — a pre-registered study testing a single specified hypothesis — the conventional threshold α = 0.05 controls the Type I error rate (false positive rate) at 5%. Here, we are doing exploratory analysis: generating hypotheses from data, not testing pre-specified ones. The goal is not to make definitive claims but to surface the strongest biological candidates for external validation in independent cohorts.

For this purpose, a threshold of p < 0.10 is a defensible choice for a candidate reporting list (see Benjamini & Hochberg, 1995, on the distinction between confirmation and exploration). It reduces the risk of missing true signals (Type II error) at the cost of a slightly elevated false-positive rate — an acceptable trade-off when every finding will require replication before any conclusion is drawn.

### Filters Applied

Pairs must satisfy all of the following:
- **Observed co-occurrence ≥ 3 runs** — minimum frequency filter to exclude singletons.
- **Empirical p-value < 0.10** — co-occurrence above the permutation null.
- **Enrichment ratio ≥ 1.0** — observed rate is not less than the null rate.
- **Mean Cramer's V ≥ 0.10** — minimum effect size filter.

The effect size filter is critical. Statistical significance and biological relevance are not the same thing: in a large enough number of permutations, nearly any non-zero difference can become statistically significant. Requiring V ≥ 0.10 ensures the association has at least a weak effect size in the contingency table — a V below 0.10 indicates negligible practical association regardless of the p-value (Cohen, 1988, benchmark: V = 0.10 small, 0.30 medium, 0.50 large).

## Part 7 — Cramer's V P-Value Pipeline (All Curious Genes)

### Complementary to the Permutation Framework

The permutation analysis in Parts 1–5 measures how often genes are *selected by the Random Forest* relative to a shuffled null. This is an indirect signal: model selection is influenced by tree depth, random subsampling, and class balance, not just co-occurrence. Part 7 takes a more direct approach — testing patient-level gene co-occurrence directly from the contingency table, without involving the classifier.

For every pair of curious genes in the class-balanced dataset, we:
1. Build a 2×2 contingency table of gene presence/absence.
2. Run **Pearson's chi-square test** (without Yates' continuity correction, per standard Cramer's V convention) to obtain χ² and its p-value.
3. Compute **Cramer's V** = √(χ²/(n·(k−1))) as an effect size measure.
4. Apply **Benjamini-Hochberg FDR correction** across all pairwise tests.

### Why Both Tests?

The permutation test and the chi-square test are sensitive to different things. The permutation test detects genes that the *model* finds useful for survival prediction; the chi-square test detects genes that *co-occur in the same patients* more than independence would predict. A gene pair can pass one test and fail the other. Using both in parallel strengthens any finding that appears in both.

### Multiple Testing

Testing all pairs of *m* genes produces *m(m−1)/2* simultaneous hypotheses. With hundreds of genes this is thousands of tests, and the probability of at least one spurious finding by chance approaches certainty. The **Benjamini-Hochberg procedure** (BH; Benjamini & Hochberg, 1995) controls the *False Discovery Rate* (FDR) — the expected fraction of reported findings that are false positives — at q = 0.05.

BH is preferred over Bonferroni (which controls the *family-wise error rate*) because Bonferroni is overly conservative when tests are positively correlated, as gene co-occurrence tests are. BH is uniformly more powerful and is the standard multiple testing correction in genomics (Storey & Tibshirani, 2003).

### References
- Benjamini, Y. & Hochberg, Y. (1995). Controlling the false discovery rate. *JRSS-B*, 57(1), 289–300.
- Storey, J.D. & Tibshirani, R. (2003). Statistical significance for genomewide studies. *PNAS*, 100(16), 9440–9445.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
from statsmodels.stats.multitest import multipletests
import warnings
warnings.filterwarnings('ignore')

# All curious-gene columns in the balanced dataset
gene_cols_all = [c for c in new_df.columns if c not in ['PATIENT_ID', 'Survived']]
gene_data_all = new_df[gene_cols_all].astype(int)
n_all = len(gene_data_all)

cramers_results = []
for i, gene_a in enumerate(gene_cols_all):
    for j, gene_b in enumerate(gene_cols_all):
        if j <= i:
            continue  # upper triangle only
        ct = pd.crosstab(gene_data_all[gene_a], gene_data_all[gene_b])
        if ct.shape != (2, 2):
            continue  # skip genes with no variation in this dataset
        chi2, p, dof, _ = chi2_contingency(ct, correction=False)
        cramers_v = np.sqrt(chi2 / (n_all * (min(ct.shape) - 1)))
        cramers_results.append({
            'Gene_A': gene_a,
            'Gene_B': gene_b,
            'Cramers_V': round(cramers_v, 6),
            'Chi2':      round(chi2,      6),
            'P_Value':   p,
        })

cramers_all_df = pd.DataFrame(cramers_results).sort_values('P_Value').reset_index(drop=True)

# Benjamini-Hochberg FDR correction
reject, p_adj, _, _ = multipletests(cramers_all_df['P_Value'], method='fdr_bh')
cramers_all_df['P_Adjusted'] = p_adj
cramers_all_df['FDR_Significant'] = reject

print(f"Total pairs tested           : {len(cramers_all_df)}")
print(f"Significant  p <= 0.05 (raw) : {(cramers_all_df.P_Value  <= 0.05).sum()}")
print(f"Significant  q <= 0.05 (FDR) : {cramers_all_df.FDR_Significant.sum()}")
print()
cramers_all_df.head(30)
```

## Part 8 — Fisher's Exact Test: Validation and Biological Interpretation

*(May 2026)*

---

### Why Chi-Squared Was Replaced

The chi-square test is the natural choice for testing association in a 2×2 contingency table. It compares observed cell counts against the counts expected under independence:

$$E_{ij} = \frac{R_i \cdot C_j}{n}$$

where R_i is the row total, C_j the column total, and n the grand total. The test is asymptotically valid when all expected cell counts exceed 5 — the rule of thumb formalised by Cochran (1954) and standard in applied statistics.

**Gene fusions are rare events.** In this 674-patient cohort, most genes are fused in only one or two patients. For a gene fused in 2 of 674 patients, the expected co-occurrence with another gene fused in 3 patients is:

$$E = \frac{2 \times 3}{674} \approx 0.009$$

This is three orders of magnitude below the minimum of 5. The chi-square approximation fails entirely here — and in practice returned p = 0.000 for pairs with **zero actual co-occurrence** in the data, a direct false positive caused by assumption violation.

**Fisher's exact test** (Fisher, 1922; extended in Fisher, 1935) solves this directly. It computes the exact hypergeometric probability of observing a table at least as extreme as the one seen, conditional on the fixed marginal totals. It requires no minimum expected cell count and is always valid for sparse 2×2 tables. It is the standard significance test for rare binary co-occurrence in biomedical genomics (Rivals et al., *Bioinformatics*, 2007).

---

### The V = 1.0 Artefact — Discard

All pairs with Cramer's V = 1.0 (χ² = 674.0) are statistical artefacts. A perfect V in a binary co-occurrence table means every patient with fusion A also has fusion B — which in genomics data almost certainly indicates: (a) the same event recorded under two different Hugo symbol designations, or (b) a gene alias collision in the source data. None of the V = 1.0 pairs carry biological meaning and all should be discarded before interpretation.

---

### Critical Caveat: Most Findings Are Driven by a Single Patient

Most significant pairs share contingency tables of the form `[[672, 0], [1, 1]]` or `[[671, 1], [1, 1]]`. This means exactly **one patient** in the balanced 674-patient dataset simultaneously carries fusions in both genes.

Fisher p = 0.003 is mathematically correct for this table structure. But the interpretation is specific: *one patient in this cohort co-carries these fusions.* This is not a population-level recurrence pattern. It may represent:

1. A **genuine genomic catastrophe phenotype** — a patient whose tumour accumulated simultaneous loss-of-function events across multiple independent tumour suppressor loci (potentially via chromothripsis; Stephens et al., *Cell*, 2011).
2. A **data quality artefact** — a sequencing or annotation error that should be verified against the clinical record before any interpretation is drawn.

Pairs supported by two or more patients (Tier 1) are substantially more credible and are prioritised below.

---

### Tier 1 — Multi-Patient Co-Occurrence (Most Robust Findings)

---

**PTEN × RB1** — V = 0.35, Fisher p = 0.012
*Table: [[669, 3], [1, 1]] — one patient carries both fusions; three carry RB1 alone.*

PTEN encodes a lipid phosphatase that directly opposes PI3-kinase, suppressing the AKT–mTOR survival axis. RB1 (retinoblastoma protein) is the master brake of the G1/S cell-cycle checkpoint: in its hypophosphorylated state, it sequesters E2F transcription factors to block S-phase entry (Dyson, 1998).

These two tumour suppressors converge on CDK activity. PTEN loss elevates cyclin D1 through the PI3K–AKT axis, activating CDK4/6. CDK4/6 phosphorylates and inactivates RB1. A patient with fusions in both genes may therefore be resistant to **both** CDK4/6 inhibitors (which require intact RB1 to re-engage the checkpoint) and PI3Kα inhibitors simultaneously — the drugs that would otherwise be the natural treatment response for each lesion individually. This co-resistance mechanism is directly supported by published clinical evidence (Juric et al., 2020; O'Brien et al., *Cancer Research*, 2020).

---

**BRCA2 × RB1** — V = 0.35, Fisher p = 0.012
*Table: [[669, 3], [1, 1]]*

BRCA2 is essential for **homologous recombination (HR)** repair of DNA double-strand breaks. BRCA2-deficient tumours are the canonical target of PARP inhibitors: PARP inhibition blocks base excision repair, creating single-strand nicks that collapse into double-strand breaks at replication forks — breaks that BRCA2-null cells cannot resolve via HR, triggering synthetic lethality (Farmer et al., *Nature*, 2005).

Concurrent RB1 loss in BRCA2-mutant triple-negative breast cancer (TNBC) is well characterised — RB1 loss promotes replication stress and may paradoxically sensitise cells further to DNA damage, but simultaneously activates alternative E2F-dependent survival programmes that can undermine PARP inhibitor response. The pipeline recovered this known co-mutation pattern without biological priors, serving as an independent validation of the approach.

---

**STK11 × RB1** — V = 0.35, Fisher p = 0.012
*Table: [[669, 3], [1, 1]]*

STK11 encodes LKB1, a serine/threonine kinase that is the primary upstream activator of AMPK — the cell's central energy sensor. LKB1–AMPK suppresses mTOR under metabolic stress and enforces cellular quiescence. STK11 loss constitutively activates mTOR, promotes aerobic glycolysis (the Warburg effect), and impairs immune surveillance by suppressing T-cell infiltration into the tumour microenvironment (Koyama et al., *Cancer Research*, 2016).

Simultaneous loss of metabolic checkpoint control (STK11) and cell-cycle checkpoint control (RB1) would give a tumour cell constitutive mTOR activation and an inoperative G1/S brake. The STK11 × RB1 co-fusion has not been characterised in breast cancer and represents a potentially novel finding.

---

**EP300 × RB1** — V = 0.35, Fisher p = 0.012
*Table: [[669, 3], [1, 1]]*

EP300 encodes the p300 histone acetyltransferase, a major transcriptional co-activator that remodels chromatin through H3K27 acetylation at enhancers and promoters. EP300 is a **direct binding partner of RB1**: pRB recruits p300/CBP complexes to repress E2F target genes during G1, and this interaction is one of the canonical mechanisms by which RB1 enforces the checkpoint at the chromatin level (Bhatt et al., *Molecular Cell*, 2004).

The recovery of this known RB1–EP300 functional relationship from co-occurrence statistics alone serves as a **positive control**: the pipeline rediscovered an established biology without being given it as input, providing confidence in the methodology.

---

**NF1 × PPM1D** — V = 0.41, Fisher p = 0.009
*Table: [[670, 1], [2, 1]] — two patients carry NF1 fusions; one co-carries PPM1D.*

NF1 encodes neurofibromin, a GTPase-activating protein (GAP) that accelerates hydrolysis of RAS-GTP to inactive RAS-GDP. NF1 loss-of-function results in **constitutive RAS activation** — functionally equivalent to a KRAS G12D point mutation in its downstream effects on MAPK/ERK and PI3K/AKT signalling, but undetectable by KRAS sequencing panels (Ratner & Miller, *Nature Reviews Cancer*, 2015).

PPM1D (Wip1 phosphatase) terminates the DNA-damage response by dephosphorylating and inactivating p53, CHK1, CHK2, and γH2AX. PPM1D amplification occurs in approximately 11% of breast cancers and represents a mechanism by which tumours disable their own damage-sensing machinery (Bulavin et al., *Nature Genetics*, 2002).

The combination is mechanistically coherent: constitutive RAS activation elevates reactive oxygen species and increases replication fork stalling, generating ongoing DNA damage. PPM1D simultaneously disables the checkpoints that would detect and respond to that damage. The tumour continuously generates genomic instability while suppressing the alarm system — a **generate-and-ignore** phenotype.

This pair is of particular clinical interest. Daraxonrasib (RMC-6236), a RAS(ON) multi-selective inhibitor, has demonstrated OS benefit in RAS-driven cancers. Breast cancer patients with NF1 fusions are functionally RAS-addicted but are currently excluded from KRAS-mutation-based eligibility criteria — a gap that fusion profiling from pipelines like this one could help close.

---

**RAD51C × PPM1D** — V = 0.41, Fisher p = 0.009
*Table: [[670, 1], [2, 1]]*

RAD51C is a member of the RAD51 paralog family and an essential component of the BRCA1/BRCA2-independent arm of homologous recombination repair. Germline RAD51C variants cause Fanconi anaemia complementation group O (FANCO) and are included in multi-gene hereditary breast and ovarian cancer panels (Loveday et al., *Nature Genetics*, 2012).

RAD51C × PPM1D is the mechanistically most striking confirmed pair in this analysis. **RAD51C cannot repair DNA double-strand breaks. PPM1D cannot detect that repair has failed.** The co-fusion creates a tumour that simultaneously lacks the repair machinery and the checkpoint designed to arrest cells when repair fails — a maximally genomically unstable phenotype. This is the molecular equivalent of driving with no brakes and no warning lights.

---

**BRIP1 × IGF1R** — V = 0.41, Fisher p = 0.009
*Table: [[670, 1], [2, 1]]*

BRIP1 (FANCJ) encodes a DNA helicase that resolves G-quadruplex secondary structures at replication forks and is required for BRCA1-mediated DNA repair. Germline BRIP1 mutations confer elevated ovarian cancer risk and are included in comprehensive hereditary cancer panels (Rafnar et al., *Nature Genetics*, 2011).

IGF1R (insulin-like growth factor 1 receptor) drives cell survival and proliferation through the PI3K/AKT and MAPK/ERK axes. High IGF1R expression correlates with poor prognosis and endocrine therapy resistance in ER-positive breast cancer. Multiple IGF1R inhibitors have been evaluated in clinical trials, though no approved agent exists in breast cancer to date.

The BRIP1 × IGF1R co-fusion has not been characterised previously — it represents a potential novel finding pairing a DNA repair deficiency with an active growth signalling vulnerability.

---

**BCAS3 × BRIP1** — V = 0.41, Fisher p = 0.009
*Table: [[670, 2], [1, 1]] — two patients carry BRIP1 fusions; one co-carries BCAS3.*

BCAS3 (Breast Cancer Amplified Sequence 3) is a transcriptional co-activator of oestrogen receptor alpha (ERα), recurrently amplified at chromosome 17q23 in luminal breast cancer. BCAS3 amplification strengthens ERα-driven transcription through a PELP1-dependent positive feedback loop, contributing to oestrogen signalling amplification.

The co-occurrence of BCAS3 (an ERα amplifier) and BRIP1 (a DNA repair gene) in the same patient suggests a tumour that is simultaneously over-activating the oestrogen receptor pathway and accumulating DNA damage through repair deficiency — a combination with potential relevance to both endocrine therapy and PARP inhibitor responsiveness.

### Tier 2 — Single-Patient Co-Occurrence Pairs (Fisher p < 0.01)

These pairs are statistically significant under Fisher's exact test but are each driven by a single patient carrying both fusions. They require clinical record verification and replication in an independent cohort before biological conclusions can be drawn.

| V | Fisher p | Notable pairs |
|---|----------|---------------|
| 0.71 | 0.003 | BRCA2–AXL, AKAP8–NOTCH3, NOTCH2–NOTCH1, PDE1C–IDH2, ETV6–FGFR1, MYC-containing pairs, RB1-containing pairs (TGFB2, HMHA1, KLHL13, AXL) |
| 0.58 | 0.004 | ACTN4–MYC, MYC–AKT2, NF1–TIMP2, ZFP64–RAD51C, ABCA8–RAD51C |

**Biologically notable single-patient pairs:**

- **BRCA2 × AXL** — AXL is a receptor tyrosine kinase that drives EMT and is upregulated in PARP inhibitor resistance. Experimental work shows BRCA2-deficient cells induce AXL-mediated EMT markers within hours of PARP inhibitor exposure. Co-fusion of BRCA2 and AXL in the same patient predicts both intrinsic PARP inhibitor sensitivity (BRCA2 loss) and a pre-existing resistance programme (AXL expression) — a potential explanation for clinical primary resistance in some BRCA2-mutant patients.

- **PDE1C × IDH2** — IDH2 mutations produce the oncometabolite 2-hydroxyglutarate (2-HG), which inhibits α-ketoglutarate-dependent dioxygenases including TET enzymes and histone demethylases, causing genome-wide epigenetic dysregulation and mTOR activation. PDE1C loss elevates cAMP/PKA signalling, a second independent growth-promoting axis. These represent two orthogonal oncogenic cascades converging simultaneously.

- **MYC × AKT2** — MYC is the master transcriptional amplifier of the proliferative programme; AKT2 is the key PI3K effector driving survival, protein synthesis via mTOR, and glucose uptake. Co-fusion in a single patient represents convergent activation of both transcriptional and kinase-signalling oncogenic programmes.

- **BRCA2 × STK11** — dual tumour suppressor co-loss across independent pathways. Both are high-penetrance susceptibility genes; their co-occurrence in breast cancer fusions has not previously been characterised.

### Collective Interpretation

The dominant pattern is not a population-wide recurrence signal. It is the identification of individual patients carrying **multiple simultaneous fusion events** across unrelated tumour suppressor and oncogene systems.

The RB1 convergence is the most striking feature. The four Tier 1 pairs involving RB1 (PTEN×RB1, BRCA2×RB1, STK11×RB1, EP300×RB1) share identical contingency tables — the same patients are simultaneously co-carrying fusions in PTEN, BRCA2, STK11, and EP300 alongside RB1. This is consistent with a **catastrophic multi-tumour-suppressor disruption phenotype**: a patient whose tumour underwent near-simultaneous loss-of-function events across the major tumour suppressor nodes, converging on RB1 as a shared downstream target. This pattern is mechanistically consistent with chromothripsis — catastrophic chromosomal shattering and reassembly in a single mitotic event — which has been documented in breast cancer (Stephens et al., *Cell*, 2011).

The most mechanistically coherent and replication-worthy findings are:
1. **NF1 × PPM1D** and **RAD51C × PPM1D** — two patients each; direct pathway logic.
2. **The RB1 convergence cluster** — multiple patients; direct literature validation.

Immediate next steps: (1) clinical record review to verify the integrity of the multi-fusion patient records, (2) replication in TCGA and METABRIC cohorts.
